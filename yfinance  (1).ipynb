```python
# Fetch Tesla stock data using the correct ticker symbol 'TSLA'
tesla = yf.Ticker('TSLA')

# Get historical data for Tesla with a maximum period
tesla_data = tesla.history(period='max')

# Display the first few rows of the Tesla stock data
```

```python
import matplotlib.pyplot as plt
tesla = yf.Ticker('TSLA')
tesla_data = tesla.history(period='1y')  # Fetch data for the last year

# Plotting the Tesla stock data (e.g., Close price)
plt.figure(figsize=(10,6))
plt.plot(tesla_data.index, tesla_data['Close'], label='Tesla Stock Close Price', color='blue')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Tesla Stock Price Over the Last Year')
plt.legend()

# Show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to make it neat
plt.show()
```
